## **삼각형 판별 자동화**

자동화는 프로그래밍 도구를 사용해서 알고리즘을 수행하는 과정입니다. 

아래 순서도를 파이썬 언어로 자동화를 하면 다음과 같습니다. 

```python
변의개수 = int(input('변의 개수를 입력하세요 : '))
각의개수 = int(input('각의 개수를 입력하세요 : '))

if 변의개수 == 3:
    if 각의개수 == 3:
        print('삼각형입니다.')
```

**변의개수**, **각의 개수**는 값을 입력받아 저장할 변수입니다.

**변의개수**와 **각의개수**가 모두 3일 때 '삼각형입니다'를 출력합니다.
